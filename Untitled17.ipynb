```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
List_Buku = []
```

```python
for page in range(1, 5):
    url = f"https://deepublishstore.com/product-category/ekonomi/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    judul = soup.find_all("h2", class_="woocommerce-loop-product__title")
    harga = soup.find_all("span", class_="woocommerce-Price-amount amount")
    gambar = soup.find_all("img", class_="attachment-woocommerce_thumbnail size-woocommerce_thumbnail")

    for i in range(len(judul)):
        data = {
            "Judul": judul[i].text.strip(),
            "Harga": harga[i].text.strip() if i < len(harga) else None,
            "Gambar": gambar[i]['src'] if i < len(gambar) else None
        }
        List_Buku.append(data)
```

```python
with open("List_Buku.json", "w", encoding="utf-8") as file:
    json.dump(List_Buku, file, ensure_ascii=False, indent=4)
```

```python
df = pd.DataFrame(List_Buku)
df.to_csv("List_Buku.csv", index=False)
```
